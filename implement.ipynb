```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
def f(x, y):
    return (x - 3.14)**2 + (y - 2.72)**2 + np.sin(3*x + 1.41) + np.sin(4*y - 1.73)
```

```python
x, y = np.array(np.meshgrid(np.linspace(0, 5, 1000), np.linspace(0, 5, 1000)))
x
```

```python
y
```

```python
z = f(x, y)
z
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z)
plt.show()
```

```python
x_min = x.ravel()[z.argmin()]
y_min = y.ravel()[z.argmin()]
```

```python
def show_figure_():
    plt.figure(figsize=(8, 6))
    plt.imshow(z, extent=[0, 5, 0, 5], origin='lower', cmap='viridis', alpha=0.5)
    plt.colorbar()
    plt.plot([x_min], [y_min], marker='x', markersize=5, color='white')
    contours = plt.contour(x, y, z, 10, colors='black', alpha=0.4)
    plt.clabel(contours, inline=True, fontsize=8, fmt='%.0f')
```

```python
def show_figure():
    fig, ax = plt.subplots(figsize=(8, 6))
    img = ax.imshow(z, extent=[0, 5, 0, 5], origin='lower', cmap='viridis', alpha=0.5)
    fig.colorbar(img, ax=ax)
    ax.plot([x_min], [y_min], marker='x', markersize=5, color='white')
    contours = ax.contour(x, y, z, 10, colors='black', alpha=0.4)
    ax.clabel(contours, inline=True, fontsize=8, fmt='%.0f')
    return fig, ax
```

```python
show_figure()
plt.show()
```

```python
n_particles = 20
# np.random.seed(100)
X = np.random.rand(2, n_particles)*5
V = np.random.randn(2, n_particles)*0.1
X
```

```python
V
```

```python
fig, ax = show_figure()
ax.scatter(x=X[0], y=X[1])
plt.show()
```

```python
pbest = X
pbest_obj = f(X[0], X[1])
gbest = pbest[:, pbest_obj.argmin()]
gbest_obj = pbest_obj.min()
```

```python
fig, ax = show_figure()
ax.scatter(x=X[0], y=X[1])
ax.scatter(x=gbest[0], y=gbest[1], marker='*')
plt.show()
```

```python
c1 = c2 = 0.1
w = 0.8

def update():
    global X, V, pbest, pbest_obj, gbest, gbest_obj
    r = np.random.rand(2)
    V = w*V + c1*r[0]*(pbest - X) + c2*r[1]*(gbest.reshape(-1, 1) - X)
    X = X + V
    obj = f(X[0], X[1])
    pbest[:, (pbest_obj >= obj)] = X[:, (pbest_obj >= obj)]
    pbest_obj = np.array([pbest_obj, obj]).min(axis=0)
    gbest = pbest[:, pbest_obj.argmin()]
    gbest_obj = pbest_obj.min()
```

```python
update()
fig, ax = show_figure()
p_plot = ax.scatter(X[0], X[1])
p_arrow= ax.quiver(X[0], X[1], V[0], V[1], angles='xy', scale_units='xy', scale=1)
pbest_plot = ax.scatter(pbest[0], pbest[1], marker='+')
gbest_plot = ax.scatter([gbest[0]], [gbest[1]], marker='*')
# pbest_plot = pbest_plot.set_offsets(pbest.T)
# p_plot = p_plot.set_offsets(X.T)
# p_arrow = p_arrow.set_offsets(X.T)
# p_arrow = p_arrow.set_UVC(V[0], V[1])
# gbest_plot = gbest_plot.set_offsets(gbest.reshape(1,-1))
plt.show()
```

```python
def animate(i):
    title = 'Iteration {:02d}'.format(i)
    update()
    ax.set_title(title)
    pbest_plot.set_offsets(pbest.T)
    p_plot.set_offsets(X.T)
    p_arrow.set_offsets(X.T)
    p_arrow.set_UVC(V[0], V[1])
    gbest_plot.set_offsets(gbest.reshape(1,-1))
    return ax, pbest_plot, p_plot, p_arrow, gbest_plot
```

```python
anim = FuncAnimation(fig, animate, frames=list(range(1, 50)), interval=500, blit=False, repeat=True)
anim.save('MyPSO.gif', dpi=120, writer='imagemagick')
```
